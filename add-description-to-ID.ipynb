```python
#Edits .gff files to append feature description to feature ID. 
```

```python
from BCBio import GFF
from BCBio.GFF import GFFExaminer

import glob
```

```python
files = glob.glob("*.gff*")
```

```python
#read all records from GFF file
examiner = GFFExaminer()

def get_records(file):
    
    in_handle = open(file)
    entries = []
    for rec in GFF.parse(in_handle):
        rec.annotations = {} #remove annotations
        entries.append(rec)
    in_handle.close()
    
    return entries
```

```python
wanted_types = ['pseudogenic_transcript',"mRNA", "protein_coding_gene", "pseudogene" ]
```

```python
for file in files:
    recs = get_records(file)
    
    for rec in recs:
        
        for feature in rec.features:
            
            t = feature.type 
            if t in wanted_types:
                #add description to iD
                desc = feature.qualifiers["description"][0].split(",")
                append = desc[0].replace(" ", "_")
                
                feature.qualifiers["ID"][0] += ":" + append
                
                print(append, feature.qualifiers["ID"][0])
            
            for subfeature in feature.sub_features:
                t = subfeature.type
                if t in wanted_types:
                    desc = subfeature.qualifiers["description"][0].split(",")
                    append = desc[0].replace(" ", "_")
                    subfeature.qualifiers["ID"][0] += ":" + append
                    print(append, subfeature.qualifiers["ID"][0])
                    
                    
    outfile = file.split(".")[0] + "_DESCRIPTION_ADDED." + file.split(".")[1]
    
    with open(outfile, "w") as f:
        GFF.write(recs, f)
```
