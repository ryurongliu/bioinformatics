```python
#Edits .gff files to append chromosome name and start/end location to feature ID. 
```

```python
from BCBio import GFF
from BCBio.GFF import GFFExaminer

import glob
```

```python
#get all gff files from current directory 
files = glob.glob("*.gff*")
```

```python
#read all records from GFF file
examiner = GFFExaminer()

def get_records(file):
    
    in_handle = open(file)
    entries = []
    for rec in GFF.parse(in_handle):
        rec.annotations = {} #remove annotations
        entries.append(rec)
    in_handle.close()
    
    return entries
```

```python
for file in files:
    recs = get_records(file)
    
    #parse all recs
    for rec in recs: 
        
        #same name for all features of one rec 
        name = rec.id
        
        #parse all features of that rec
        for feature in rec.features:
            
#             print(feature.qualifiers['ID'])
            #get start and end points 
            start = str(int(feature.location.start) + 1)
            end = str(feature.location.end)
             
            #change feature ID
            feature.qualifiers['ID'][0] += ":" + name + ":" + start + "_" + end

            for subfeature in feature.sub_features:
                start = str(int(subfeature.location.start) + 1)
                end = str(subfeature.location.end)
                subfeature.qualifiers['ID'][0] += ":" + name + ":" + start + "_" + end
                
                for sub2 in subfeature.sub_features:
                    start = str(int(sub2.location.start) + 1)
                    end = str(sub2.location.end)
                    sub2.qualifiers['ID'][0] += ":" + name + ":" + start + "_" + end
                
            
    #write to outfile 
    outfile = file.split(".")[0] + "_ID_EDITED." + file.split(".")[1]
    with open(outfile, "w") as f:
        GFF.write(recs, f)
```
